```python
#App with highest downloads
def HeighDownloadApp(df):
    k=max(list(dict(df.rating_count_tot.value_counts()).keys()))
#     print(k)
    for i in range(len(df['track_name'])):
        if df['rating_count_tot'][i]==k:
            return df['track_name'][i]
HeighDownloadApp(android)
```

### Average of rating for free apps

```python
#Average user rating for free apps
def AverageFreeAppRating(df):
    k=df.user_rating.values
    count=l=0
    for i in range(len(k)):
        if android['price'][i]==0:
            count+=k[i]
            l+=1
    return count/l
AverageFreeAppRating(android)
```

### Average user rating for paid apps

```python
#Average user rating for paid apps
def AveragePaidAppRating(df):
    k=df.user_rating.values
    count=l=0
    for i in range(len(k)):
        if df['price'][i]!=0:
            count+=k[i]
            l+=1
    return count/l
AveragePaidAppRating(android)
```

```python
s.countplot(y="user_rating",data=android)
p.show()
```

### Category with highest average user rating for paid apps

```python
#Category with highest average user rating for paid apps
def highRatingPaidCategory(android):
    li=[]
    a=android['prime_genre'].unique()
    for j in a:
        k=android.user_rating.values
        count=l=0
        for i in range(len(k)):
            if android['price'][i]!=0 and android['prime_genre'][i]==j:
                count+=k[i]
                l+=1
        li.append(count/l)
    for i in range(len(li)):
        if li[i]==max(li):
            print(a[i])
    return
highRatingPaidCategory(android)
```

### Compare average user rating for paid vs free gaming apps

```python
#Compare average user rating for paid vs free gaming apps
def compare(android):
    p=AveragePaidAppRating(android)
    k=android.user_rating.values
    count=l=0
    for i in range(len(k)):
        if android['price'][i]==0 and android['prime_genre'][i]=='Games':
            count+=k[i]
            l+=1
    if (count/l)>p:
        print('Average user_rating free gaming apps')
    else:
        print('Average user_raring paid apps')
compare(android)
```
